```python
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic')
size = [2441, 2312, 1031, 1233] # 각 조각에 해당하는 레이블 사이즈
label = ['A형', 'B형', 'O형', 'AB형']
plt.axis('equal') #그래프 축 비율을 동일하게 설정
plt.pie(size, labels = label)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic')
size = [2441, 2312, 1031, 1233]
label = ['A형', 'B형', 'O형', 'AB형']
plt.axis('equal')
plt.pie(size, labels = label, autopct='%.1f%%')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic')
size = [2441, 2312, 1031, 1233]
label = ['A형', 'B형', 'O형', 'AB형']
color = ['skyblue', 'steelblue', 'powderblue', 'aliceblue']
plt.axis('equal')
plt.pie(size, labels = label, autopct='%.1f%%', colors = color, explode=(0,0,0.1,0))
plt.legend()
plt.show()
```
