```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)
result = []

for row in data:
    if row[-1] != '':
        result.append(float(row[-1]))

plt.hist(result, bins = 100, color='r')
plt.show()
```

```python
import csv

f = open('seoul.csv')
data = csv.reader(f)
next(data)
rs = []

for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '08':
            rs.append(float(row[-1]))

plt.hist(rs, bins=100)
plt.show()
```

```python
import csv

f = open('seoul.csv')
data = csv.reader(f)
next(data)
r1 = []
r2 = []

for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month == '08':
            r1.append(float(row[-1]))
        if month == '01':
            r2.append(float(row[-1]))

plt.hist(r1, bins=100, color='r', label='Aug')
plt.hist(r2, bins=100, color='b', label='Jan')
plt.legend()
plt.show()
```

```python
#8월의 기온 데이터를 추출, 날짜별 기온의 분포
#박스 플롯: 각 날짜에 대한 기온의 분포&이상치를 시각적으로 보여줌
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)

day=[]

for i in range(31):
    day.append([])

for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '08':
            day[int(row[0].split('-')[2])-1].append(float(row[-1]))

plt.figure(figsize=(10,5), dpi=300)
plt.boxplot(day)
plt.show
```
