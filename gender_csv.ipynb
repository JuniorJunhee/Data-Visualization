```python
import csv
f = open('gender.csv')
data = csv.reader(f)

m = [] #남자
f = [] #여자

for row in data:
    if '신도림'in row[0]: #행정구역이 신도림이 들어가있다면
        for i in range(0, 101): #101번 반복
            m.append(-int(row[i+3]))
            f.append(int(row[-(i+1)]))
f.reverse()

import matplotlib.pyplot as plt
plt.barh(range(101), m, color = 'b', label='남성')
plt.barh(range(101), f, color = 'r', label='여성')
plt.rcParams['axes.unicode_minus'] = False
plt.rc('font', family="Malgun Gothic")
plt.legend() #레이블(행주) 표시
plt.show()
```

```python
import csv
f = open('gender.csv')
data = csv.reader(f)
name = input('지역을 입력해주세요 : ')

m = [] #남자
f = [] #여자

for row in data:
    if name in row[0]: #행정구역이 신도림이 들어가있다면
        for i in range(0, 101): #101번 반복
            m.append(-int(row[i+3].replace(',','')))
            f.append(int(row[-(i+1)].replace(',','')))
        break
f.reverse()
print(len(m), len(f))
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.figure(figsize=(19,5), dpi=300)
plt.barh(range(101), m, color = 'b', label='남성')
plt.barh(range(101), f, color = 'r', label='여성')
plt.title(name+'지역의 남녀 성별 인구분포')
plt.rcParams['axes.unicode_minus'] = False
plt.rc('font', family="Malgun Gothic")
plt.legend() #레이블(행주) 표시
plt.show()
```

```python
import csv
f = open('gender.csv')
data = csv.reader(f)
size = []
name = input('찾고싶은 지역의 이름을 알려주세요: ')

for row in data:
    if name in row[0]:
        m = 0
        f = 0
        for i in range(101):
            m += int(row[i+3].replace(',',''))
            f += int(row[i+106].replace(',',''))
        break
size.append(m)
size.append(f)
print(size)

import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic')
color = ['pink', 'skyblue']
plt.axis('equal')
plt.pie(size, labels = ['여자', '남자'], autopct='%.1f%%', colors = color, startangle = 90)
plt.title(name+'의 남녀 비율')
plt.legend()
plt.show()
```
